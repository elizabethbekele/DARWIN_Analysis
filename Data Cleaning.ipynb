### Subset of tasks 

Inspect 5 of the 25 tasks to see how effective they were at classifying participants as healthy or with Alzheimer's.

- **Task 3:** join two points with a vertical line, continuously for four times (g)
- **Task 8:**  write cursively a sequence of four lowercase letter I in a single smooth movement (c)
- **Task 13:** copy the word “mamma” above a line (c)
- **Task 23:** write a telephone number under dictation. (m)

```python
# ID column 
# Identifies each participant
id_col = darwin['ID']
id_col[0:5]
```

```python
# Class column 
# P = patient (has Alzheimer's)
# H = healthy
class_col = darwin['class']
class_col[0:5]
```

```python
# Task 3, 13, 23
three = [col for col in darwin if col.endswith('3')]
three[0:5]
```

```python
# Task 8 & 18
eight = [col for col in darwin if col.endswith('8')]
eight[0:5]
```

```python
# Combine the 5 tasks
tasks = three + eight

# First 5 columns reference task 3
tasks[:5]
```

```python
# last 5 columns reference task 18
tasks[85:]
```

```python
# Subset entire dataframe with tasks
darwin[tasks].head()
```

```python
# Create new dataframe
# Concat tasks, class_col, id_col into new dataframe
darwin2 = pd.concat([id_col, darwin[tasks], class_col], axis = 1)
darwin2.head()
```

```python
# Export new dataframe as a csv
darwin2.to_csv('darwin2.csv',index=False)
```
